```python
def create_dataset():
    # Define feature names
    names = ['filepath','label','hu[0]','hu[1]','hu[2]','hu[3]','hu[4]','hu[5]','hu[6]','entropy','inverse_diff_moments','contrast',
             'dissimilarity','homogeneity', 'ASM','energy','correlation','mean_h','mean_s','mean_v', 'stddev_h','stddev_s',
             'stddev_v','hog_feature']

    df = pd.DataFrame([], columns=names)
    for i, directory in enumerate(os.listdir('Leaf')):
        for filename in os.listdir(os.path.join('Leaf', directory)):
            # Reading the image
            img = cv2.imread(os.path.join('Leaf', directory, filename))
            
            # Get the full file path
            filepath = os.path.join('Leaf', directory, filename)

            # Resize image to 100x100
            image = cv2.resize(img, (100, 100))

            # Apply median filter
            image = cv2.medianBlur(image, 3)

            # Apply Gaussian filter
            image = cv2.GaussianBlur(image, (5, 5), 0)

            # Convert image to grayscale
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Vein Segmentation
            threshold = filters.threshold_otsu(gray)
            binary = gray > threshold

            # Vein Enhancement (Optional)
            kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
            enhanced_binary = cv2.morphologyEx(binary.astype(np.uint8), cv2.MORPH_OPEN, kernel)

            # Vein Extraction
            skeleton = cv2.ximgproc.thinning(enhanced_binary)

            # Estimate the noise standard deviation
            sigma_est = np.mean(estimate_sigma(skeleton, channel_axis=None))

            # Perform non-local means denoising
            skeleton = (denoise_nl_means(skeleton, h=1.15 * sigma_est, fast_mode=False, patch_size=5, patch_distance=3) * 255).astype(
                np.uint8)

            # Convert image to HSV
            hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

            # Split the channels
            h, s, v = cv2.split(hsv)

            # Calculate Hu moments
            moments = cv2.moments(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
            hu = cv2.HuMoments(moments).flatten()

            # Calculate the GLCM properties
            glcm = graycomatrix(skeleton, [1, 2], [0, np.pi / 4, np.pi / 2, 3 * np.pi / 4], levels=256, symmetric=True, normed=True)
            contrast = graycoprops(glcm, 'contrast').mean()
            dissimilarity = graycoprops(glcm, 'dissimilarity').mean()
            homogeneity = graycoprops(glcm, 'homogeneity').mean()
            ASM = graycoprops(glcm, 'ASM').mean()
            energy = graycoprops(glcm, 'energy').mean()
            correlation = graycoprops(glcm, 'correlation').mean()

            # Calculate the mean and standard deviation of each channel
            mean_h = np.mean(h) if h.size > 0 else 0.0
            mean_s = np.mean(s) if s.size > 0 else 0.0
            mean_v = np.mean(v) if v.size > 0 else 0.0
            stddev_h = np.std(h) if h.size > 0 else 0.0
            stddev_s = np.std(s) if s.size > 0 else 0.0
            stddev_v = np.std(v) if v.size > 0 else 0.0

            # Calculate the HOG feature
            hog_feature = hog(skeleton, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), block_norm='L2-Hys')[0]

            # Haralick features
            textures = mahotas.features.haralick(skeleton)
            ht_mean = textures.mean(axis=0)
            inverse_diff_moments = ht_mean[4]
            entropy = ht_mean[8]
            label=i
            
            # Feature vector
            vector = [
                filepath,
                label,
                hu[0],hu[1],hu[2],hu[3],hu[4],hu[5],hu[6],
                entropy,
                inverse_diff_moments,
                contrast,
                dissimilarity,
                homogeneity,
                ASM,
                energy,
                correlation,
                mean_h,
                mean_s,
                mean_v,
                stddev_h,
                stddev_s,
                stddev_v,
                hog_feature
            ]
            df_temp = pd.DataFrame([vector],columns=names)
            df = pd.concat([df, df_temp], ignore_index=True)
            print(filename)
            s=[i,100,100,3]
            images.append(vector)
            labels.append(i)
    return df
```

# Split the data into training and testing sets

```python
dataset = pd.read_csv('features1.csv')
```

```python
train, test = train_test_split(dataset, test_size=0.2, random_state=0)
```

```python
# Define image generators
train_datagen = ImageDataGenerator(rescale=1./255,
rotation_range=30,
zoom_range=0.2,
horizontal_flip=True,
vertical_flip=True,
fill_mode='nearest')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train['label'] = train['label'].astype(str)
```

```python
# Flow training images in batches of 32 using train_datagen generator
train_generator = train_datagen.flow_from_dataframe(
train,
x_col='filepath',
y_col='label',
target_size=(100, 100),
batch_size=32,
class_mode='categorical',
shuffle=True
)
```

```python
test['label'] = test['label'].astype(str)
```

```python
# Flow validation images in batches of 32 using test_datagen generator
test_generator = test_datagen.flow_from_dataframe(
test,
x_col='filepath',
y_col='label',
target_size=(100, 100),
batch_size=32,
class_mode='categorical',
shuffle=True
)
```

```python
# Create model

model = Sequential()

# Add convolutional layers

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(100, 100, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
# Add dropout
model.add(Dropout(0.5))
# Add flatten layer
model.add(Flatten())
# Add dense layers
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

# Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()

# Set the number of epochs
num_epochs = 100

# Lists to store loss and accuracy values for each epoch
train_loss = []
train_acc = []
val_loss = []
val_acc = []

# Get the number of steps per epoch
steps_per_epoch = len(train_generator)

# Training loop
for epoch in range(num_epochs):
    # Reset metrics at the start of each epoch
    model.reset_metrics()
    
    # Variables to accumulate loss and accuracy over steps
    epoch_loss = 0.0
    epoch_acc = 0.0
    
    # Iterate over the training data in batches
    for step in range(steps_per_epoch):
        # Generate batch data
        batch_data, batch_labels = next(train_generator)
        
        # Perform a forward pass
        predictions = model.train_on_batch(batch_data, batch_labels)
        
        # Accumulate the loss and accuracy
        epoch_loss += predictions[0]
        epoch_acc += predictions[1]
    
    # Calculate the average loss and accuracy for the epoch
    avg_loss = epoch_loss / steps_per_epoch
    avg_acc = epoch_acc / steps_per_epoch
    
    # Evaluate the model on the validation dataset
    val_metrics = model.evaluate(test_generator)
    val_loss.append(val_metrics[0])
    val_acc.append(val_metrics[1])
    
    # Store the average loss and accuracy
    train_loss.append(avg_loss)
    train_acc.append(avg_acc)
    
    # Print the metrics for the epoch
    print(f'Epoch {epoch+1}/{num_epochs} - Training Loss: {avg_loss:.4f} - Training Accuracy: {avg_acc:.4f} - Validation Loss: {val_metrics[0]:.4f} - Validation Accuracy: {val_metrics[1]:.4f}')
```

```python
# Plot the training and validation loss and accuracy curves
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Plot the smoothed training loss
axs[0].plot(train_loss, label='Training Loss')
axs[0].plot(val_loss, label='Validation Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
axs[0].legend()

# Plot the smoothed training accuracy
axs[1].plot(train_acc, label='Training Accuracy')
axs[1].plot(val_acc, label='Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()

plt.show()
```
